````python
def copilot_proportion_circle():
    copilot_users = 3_450_113
    total_developers = 7.5 * 10**6
    proportion = copilot_users / total_developers
    print(f'The proportion of developers using Copilot is {proportion:.2%} or {copilot_users} out of {total_developers} developers.')
    # let's add in a constraint for our data visual that the total number of developers is 100.
    total_developers = 500 # increase the number of total developers
    copilot_users = int(proportion * total_developers) # we need to round this number to the nearest integer
    
    fig, ax = plt.subplots(figsize=(20, 20), subplot_kw={'aspect': 'equal'})

    copilot_icon = Image.open(request.urlopen("https://github.githubassets.com/images/icons/copilot/cp-head-square.png"))
    copilot_icon = np.array(copilot_icon)
    copilot_icon = OffsetImage(copilot_icon, zoom=0.03 * np.sqrt(total_developers)/10) # smaller size proportional to total_developers

    not_copilot_icon = Image.open("./images/user.png") # load the blank user icon
    not_copilot_icon = np.array(not_copilot_icon)
    not_copilot_icon = OffsetImage(not_copilot_icon, zoom=0.06 * np.sqrt(total_developers)/10) # smaller size proportional to total_developers

    copilot_developers = proportion * total_developers
    not_copilot_developers = total_developers - copilot_developers

    # Place the icons along a circle
    angle = np.linspace(0, 2 * np.pi, total_developers, endpoint=False)
    radius = np.ones(total_developers)
    x = radius * np.cos(angle)
    y = radius * np.sin(angle)
    for i in range(int(copilot_developers)):
        ax.add_artist(offsetbox.AnnotationBbox(copilot_icon, [x[i], y[i]], frameon=False))
    for i in range(int(not_copilot_developers)):
        ax.add_artist(offsetbox.AnnotationBbox(not_copilot_icon, [x[i + int(copilot_developers)], y[i + int(copilot_developers)]], frameon=False))

    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.axis('off')
    plt.title("Proportion of Copilot and Non-Copilot Developers")
    # save the image as a png named copilot_proportion_circle.png in the images folder
    plt.savefig('./images/copilot_proportion_circle.png')

copilot_proportion_circle()
````
